```python
%%bash

mkdir -p src
mkdir -p build
```

```python
%%writefile CMakeLists.txt

# Specify the minimum required version of CMake
cmake_minimum_required(VERSION 3.12)

# Set the project name and version
project(MyProject VERSION 1.0)

# Define a variable for the program name
set(PROGRAM_NAME main)

# Add compile options
add_compile_options(-std=c++20 -Wall -Wextra -Werror)

# Create an executable target using the variable
add_executable(${PROGRAM_NAME} src/main.cpp)

# Diagnostic messages
message(STATUS "Compiler ID: ${CMAKE_CXX_COMPILER_ID}")
message(STATUS "Compiler Version: ${CMAKE_CXX_COMPILER_VERSION}")
message(STATUS "Compiler Path: ${CMAKE_CXX_COMPILER}")

# After defining target, you can print its properties like link libraries
get_target_property(LINKED_LIBRARIES ${PROGRAM_NAME} LINK_LIBRARIES)
if(LINKED_LIBRARIES)
    message(STATUS "Libraries linked to ${PROGRAM_NAME}: ${LINKED_LIBRARIES}")
else()
    message(STATUS "No libraries linked to ${PROGRAM_NAME}.")
endif()
```

```python
%%cpp -n src/main.cpp

#include <iostream>
#include <vector>
#include <string>

void foo(const auto& a, const auto& b) {
  std::cout << a << " " << b << "\n";
}

auto main() -> int {
  foo(1, std::string("Hello!"));
  foo(std::string("Hello"), 42);
  return 0;
}
```

```python
%%bash

cd build
cmake --fresh ..
make
./main
```
